### Overview of the Iris Dataset Analysis

This Jupyter Notebook provides a comprehensive analysis of the Iris dataset using various Python libraries such as `pandas`, `numpy`, `seaborn`, and `matplotlib`. Below is a summary of the key steps and topics covered in this notebook:

1. **Importing Libraries**:
    - The notebook starts by importing essential libraries including `pandas`, `numpy`, `seaborn`, and `matplotlib` for data manipulation and visualization.
    - The `sklearn` library is also imported for accessing the Iris dataset.

2. **Loading the Iris Dataset**:
    - The Iris dataset is loaded using `load_iris` from the `sklearn.datasets` module.
    - The dataset is then converted into a pandas DataFrame for easier manipulation and analysis.

3. **Exploratory Data Analysis (EDA)**:
    - The shape of the DataFrame is checked to understand the dimensions of the dataset.
    - The first few rows of the DataFrame are displayed using the `head()` method.
    - Descriptive statistics of the dataset are generated using the `describe()` method.
    - The unique values in each column are counted using the `nunique()` method.

4. **Data Cleaning**:
    - A new column `target` is added to the DataFrame to include the target labels from the Iris dataset.
    - A cleaned version of the DataFrame is created by dropping the `target` column.

5. **Correlation Analysis**:
    - The correlation matrix of the cleaned DataFrame is computed using the `corr()` method.
    - Heatmaps of the correlation matrix are generated using `seaborn` to visualize the relationships between different features.

6. **Data Visualization**:
    - Scatter plots are created to visualize the relationships between different pairs of features in the dataset.
    - The `seaborn` library is used to create scatter plots with additional information such as hue to differentiate between different values.

### Topics Covered:
- Data Loading and Preprocessing
- Exploratory Data Analysis (EDA)
- Data Cleaning
- Correlation Analysis
- Data Visualization using `matplotlib` and `seaborn`

This notebook provides a solid foundation for understanding the Iris dataset and demonstrates various techniques for data analysis and visualization.

### Import the libs  

```python
import sklearn as sklearn   
import pandas as pd
import numpy as np 
import matplotlib as mp 
import seaborn as sns
```


###  import the data from Sklearn

```python
from sklearn.datasets import load_iris  
```

```python
iris_dataset = load_iris()  
```

```python
dataframe  = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)    
```

```python
dataframe.shape
```

### View and Analyse the data

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
dataframe.columns
```

```python
dataframe.nunique(axis=0)
```

```python

iris_dataset.target_names
```

```python
dataframe['target'] = iris_dataset.target
```

```python
dataframe  
```

```python
df_cleaned = dataframe.drop(['target' ],axis=1)
```

```python
df_cleaned
```

## Visualise the dataset
#### SNS heatmap docs : https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
corr = df_cleaned.corr()
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
sns.heatmap(corr,cmap="crest",annot=True,linewidths="5")
```

### some helfpul links to change your colours : https://www.python-graph-gallery.com/92-control-color-in-seaborn-heatmaps


### nice built in ones : https://seaborn.pydata.org/tutorial/color_palettes.html

```python
sns.heatmap(corr, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
df_cleaned.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)')
```

```python
df_cleaned.plot(kind='scatter', x = 'petal length (cm)', y = 'petal width (cm)')
```

```python
sns.scatterplot(data=df_cleaned, x="sepal width (cm)", y="sepal length (cm)", hue="sepal width (cm)")
```
